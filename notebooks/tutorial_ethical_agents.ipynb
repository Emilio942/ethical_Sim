```python
# 📦 Import aller benötigten Module
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import json

# Import der Simulation-Module
from neural_society import NeuralEthicalSociety
from agents import NeuralEthicalAgent
from scenarios import ScenarioGenerator
from metrics import MetricsCollector  
from validation import ValidationSuite
from export_reporting import DataExporter, AutomatedReporter
from visualization import EthicalSimulationVisualizer

# Styling für bessere Plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("✅ Alle Module erfolgreich importiert!")
print("🚀 Bereit für die Simulation!")
```

## 🤖 Schritt 1: Ethische Agenten erstellen

Agenten sind die Kernkomponenten unserer Simulation. Jeder Agent hat:
- **Persönlichkeit**: Bestimmt Entscheidungstendenzen
- **Ethisches Framework**: Grundlegende ethische Prinzipien  
- **Lernmechanismen**: Wie der Agent aus Erfahrungen lernt
- **Soziale Eigenschaften**: Wie er mit anderen interagiert

### Verfügbare Persönlichkeitstypen:
- `utilitarian` - Maximiert Gesamtnutzen
- `deontological` - Folgt festen Regeln und Pflichten
- `virtue_ethics` - Strebt nach Tugenden und Charakterstärke
- `balanced` - Ausgewogener Mix aller Ansätze
- `pragmatic` - Praktische, situative Entscheidungen
- `idealistic` - Strebt nach höchsten Idealen

```python
# 🏗️ Erstelle eine diverse Gruppe von Agenten
print("🏗️ Erstelle ethische Agenten...")

# Verschiedene Agent-Konfigurationen
agent_configs = [
    {"id": "alice", "personality": "utilitarian", "framework": "utilitarian"},
    {"id": "bob", "personality": "deontological", "framework": "deontological"}, 
    {"id": "charlie", "personality": "virtue_ethics", "framework": "virtue_ethics"},
    {"id": "diana", "personality": "balanced", "framework": "balanced"},
    {"id": "eve", "personality": "pragmatic", "framework": "pragmatic"},
    {"id": "frank", "personality": "idealistic", "framework": "idealistic"}
]

# Agenten erstellen
agents = []
for config in agent_configs:
    agent = NeuralEthicalAgent(
        agent_id=config["id"],
        personality=config["personality"],
        ethical_framework=config["framework"]
    )
    agents.append(agent)
    print(f"✅ {config['id']}: {config['personality']} Agent erstellt")

print(f"\n🎉 {len(agents)} Agenten erfolgreich erstellt!")

# Zeige Agent-Details
for agent in agents:
    print(f"🤖 {agent.agent_id}: {agent.personality} ({agent.ethical_framework})")
    print(f"   Überzeugungen: {len(agent.beliefs.beliefs)} Beliefs")
    print(f"   Entscheidungen: {len(agent.decision_history)}")
    print()
```

## 🏛️ Schritt 2: Gesellschaft erstellen und Simulation durchführen

Die `NeuralEthicalSociety` koordiniert Interaktionen zwischen Agenten:
- **Soziale Netzwerke**: Dynamische Verbindungen zwischen Agenten
- **Gruppenbildung**: Automatische Community-Erkennung
- **Einflussverteilung**: Leadership und Meinungsführerschaft
- **Lernprozesse**: Sozialer Einfluss und Wissensaustausch

```python
# 🏛️ Gesellschaft erstellen
print("🏛️ Erstelle ethische Gesellschaft...")
society = NeuralEthicalSociety()

# Agenten zur Gesellschaft hinzufügen  
for agent in agents:
    society.add_agent(agent)
    print(f"➕ {agent.agent_id} zur Gesellschaft hinzugefügt")

print(f"✅ Gesellschaft mit {len(society.agents)} Agenten erstellt")

# 🎬 Szenarien generieren und durchführen
print("\n🎬 Generiere ethische Szenarien...")
scenario_gen = ScenarioGenerator()

# Mix aus vordefinierten und zufälligen Szenarien
scenarios = [
    scenario_gen.get_trolley_problem(),
    scenario_gen.get_autonomous_vehicle_dilemma(), 
    scenario_gen.get_privacy_vs_security_dilemma()
]

# Zusätzliche zufällige Szenarien
for i in range(7):
    scenarios.append(scenario_gen.generate_random_scenario())

print(f"📝 {len(scenarios)} Szenarien generiert")

# Simulation durchführen
print("\n🚀 Starte Simulation...")
for i, scenario in enumerate(scenarios, 1):
    print(f"  📝 Scenario {i}/{len(scenarios)}: {scenario.title}")
    decisions = society.run_scenario(scenario)
    print(f"    ✅ {len(decisions)} Entscheidungen getroffen")

print("🎉 Simulation abgeschlossen!")
```

## 📊 Schritt 3: Umfassende Metriken sammeln

Das Metriken-System analysiert verschiedene Aspekte der Simulation:

### 🏛️ Gesellschaftliche Metriken:
- **Polarisierung**: Wie stark gespalten sind die Meinungen?
- **Konsens**: Wie einig sind sich die Agenten? 
- **Netzwerk-Kohäsion**: Wie zusammenhängend ist das soziale Netzwerk?
- **Einflussverteilung**: Wer hat wie viel Einfluss?

### 🤖 Agenten-Metriken:
- **Interne Konsistenz**: Sind die Überzeugungen widerspruchsfrei?
- **Zeitliche Stabilität**: Wie stabil sind Überzeugungen über Zeit?
- **Entscheidungsqualität**: Wie selbstsicher und durchdacht sind Entscheidungen?
- **Lernfortschritt**: Wie effektiv lernt der Agent?

```python
# 📊 Metriken sammeln
print("📊 Sammle umfassende Metriken...")
metrics_collector = MetricsCollector()
all_metrics = metrics_collector.collect_all_metrics(society)

print("📈 Gesellschaftliche Metriken:")
societal_metrics = all_metrics.get('societal_metrics', {})
for metric, value in societal_metrics.items():
    if isinstance(value, (int, float)):
        print(f"  🏛️ {metric}: {value:.3f}")
    else:
        print(f"  🏛️ {metric}: {value}")

print("\n👥 Agenten-Metriken (Durchschnitt):")
agent_metrics = all_metrics.get('agent_metrics', {})
if agent_metrics:
    # Berechne Durchschnittswerte
    avg_metrics = {}
    for agent_id, metrics in agent_metrics.items():
        for metric, value in metrics.items():
            if isinstance(value, (int, float)):
                if metric not in avg_metrics:
                    avg_metrics[metric] = []
                avg_metrics[metric].append(value)
    
    for metric, values in avg_metrics.items():
        avg_value = np.mean(values)
        print(f"  🤖 {metric}: {avg_value:.3f}")

print(f"\n✅ {len(all_metrics)} Kategorien von Metriken gesammelt!")

# Zeige Schlüssel-Insights
print("\n🔍 Schlüssel-Insights:")
polarization = societal_metrics.get('polarization', 0)
consensus = societal_metrics.get('consensus', 0) 
cohesion = societal_metrics.get('network_cohesion', 0)

print(f"📈 Polarisierung: {polarization:.3f} ({'Niedrig' if polarization < 0.3 else 'Hoch' if polarization > 0.7 else 'Mittel'})")
print(f"🤝 Konsens: {consensus:.3f} ({'Niedrig' if consensus < 0.3 else 'Hoch' if consensus > 0.7 else 'Mittel'})")
print(f"🕸️ Netzwerk-Kohäsion: {cohesion:.3f} ({'Schwach' if cohesion < 0.3 else 'Stark' if cohesion > 0.7 else 'Mittel'})")
```

## 🔍 Schritt 4: Validierung und Qualitätssicherung

Das Validierungssystem überprüft:
- **Agent-Konsistenz**: Sind Überzeugungen und Entscheidungen logisch konsistent?
- **Gesellschafts-Stabilität**: Ist das soziale Netzwerk stabil und realistisch?
- **Entscheidungsplausibilität**: Sind die Entscheidungen nachvollziehbar?
- **Lernvalidität**: Lernen die Agenten in sinnvoller Weise?
- **Anomalie-Erkennung**: Gibt es ungewöhnliche oder problematische Muster?

```python
# 🔍 Validierung durchführen
print("🔍 Führe umfassende Validierung durch...")
validator = ValidationSuite()
validation_results = validator.validate_society(society)

# Zeige Validierungsergebnisse  
print("📋 Validierungsergebnisse:")
passed = validation_results.get('passed_checks', 0)
total = validation_results.get('total_checks', 0) 
score = validation_results.get('overall_score', 0)

print(f"  ✅ Bestandene Tests: {passed}/{total}")
print(f"  📊 Gesamtscore: {score:.1%}")

# Detaillierte Ergebnisse nach Kategorie
results_by_type = validation_results.get('results_by_type', {})
print("\n📂 Ergebnisse nach Validierungstyp:")
for val_type, type_results in results_by_type.items():
    type_passed = type_results.get('passed', 0)
    type_total = type_results.get('total', 0)
    type_score = type_results.get('average_score', 0)
    print(f"  🔸 {val_type}: {type_passed}/{type_total} ({type_score:.1%})")

# Zeige Warnungen und Probleme
all_results = validation_results.get('all_results', [])
warnings = [r for r in all_results if r.get('severity') == 'warning']
errors = [r for r in all_results if r.get('severity') == 'error']

if warnings:
    print(f"\n⚠️ Warnungen ({len(warnings)}):")
    for warning in warnings[:3]:  # Nur erste 3 zeigen
        print(f"  ⚠️ {warning.get('check_name', 'Unknown')}: {warning.get('message', 'No message')}")
    if len(warnings) > 3:
        print(f"  ... und {len(warnings) - 3} weitere Warnungen")

if errors:
    print(f"\n❌ Fehler ({len(errors)}):")
    for error in errors[:3]:  # Nur erste 3 zeigen
        print(f"  ❌ {error.get('check_name', 'Unknown')}: {error.get('message', 'No message')}")
    if len(errors) > 3:
        print(f"  ... und {len(errors) - 3} weitere Fehler")

print(f"\n✅ Validierung abgeschlossen! Score: {score:.1%}")
```

## 📈 Schritt 5: Interaktive Visualisierungen

Visualisierungen helfen beim Verstehen der komplexen Dynamiken:
- **Netzwerk-Grafiken**: Soziale Verbindungen und Einfluss
- **Belief-Evolution**: Wie sich Überzeugungen über Zeit entwickeln
- **Entscheidungsmatrizen**: Muster in ethischen Entscheidungen
- **Dashboard-Übersicht**: Zusammenfassung aller wichtigen Metriken

```python
# 📈 Visualisierungen erstellen
print("📈 Erstelle Visualisierungen...")
viz = EthicalSimulationVisualizer(society)

# 1. Dashboard mit allen wichtigen Plots
plt.figure(figsize=(15, 10))
viz.create_dashboard()
plt.suptitle("🎯 Ethische Agenten-Simulation Dashboard", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print("✅ Dashboard erstellt!")
```

```python
# 2. Detaillierte Einzelvisualisierungen
print("\n📊 Erstelle detaillierte Einzelplots...")

# Netzwerk-Visualisierung
plt.figure(figsize=(12, 8))
viz.plot_network()
plt.title("🕸️ Soziales Netzwerk der Agenten", fontsize=14, fontweight='bold')
plt.show()

# Belief-Evolution über Zeit
plt.figure(figsize=(12, 6))
viz.plot_belief_evolution()
plt.title("🔄 Evolution der Überzeugungen", fontsize=14, fontweight='bold')
plt.show()

# Entscheidungsverteilung
plt.figure(figsize=(10, 8))
viz.plot_decision_distribution()
plt.title("🎯 Entscheidungsverteilung", fontsize=14, fontweight='bold')
plt.show()

print("✅ Alle Visualisierungen erstellt!")
```

## 💾 Schritt 6: Datenexport und Reporting

Das Export-System unterstützt verschiedene Formate für weitere Analyse:
- **JSON**: Strukturierte Daten für Programmierung
- **CSV**: Tabellendaten für Excel/Pandas 
- **SQLite**: Relationale Datenbank für komplexe Abfragen
- **HTML**: Interaktive Berichte für Präsentationen
- **Markdown**: Dokumentation und Reports

```python
# 💾 Export und Reporting
print("💾 Erstelle Exports und Berichte...")
exporter = DataExporter(society)

# Timestamp für eindeutige Dateinamen
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

# 1. JSON Export (strukturierte Daten)
json_file = f"tutorial_export_{timestamp}.json"
exporter.export_json(json_file)
print(f"✅ JSON Export erstellt: {json_file}")

# 2. CSV Export (Tabellendaten)
csv_file = f"tutorial_export_{timestamp}.csv" 
exporter.export_csv(csv_file)
print(f"✅ CSV Export erstellt: {csv_file}")

# 3. HTML Bericht (interaktiv)
reporter = AutomatedReporter(society)
html_file = f"tutorial_report_{timestamp}.html"
reporter.generate_html_report(html_file)
print(f"✅ HTML Bericht erstellt: {html_file}")

# 4. Zusammenfassung erstellen
summary = {
    'simulation_timestamp': timestamp,
    'agents_count': len(society.agents),
    'scenarios_count': len(scenarios),
    'total_decisions': sum(len(agent.decision_history) for agent in society.agents),
    'key_metrics': {
        'polarization': societal_metrics.get('polarization', 0),
        'consensus': societal_metrics.get('consensus', 0),
        'network_cohesion': societal_metrics.get('network_cohesion', 0)
    },
    'validation_score': validation_results.get('overall_score', 0),
    'files_created': [json_file, csv_file, html_file]
}

# Zusammenfassung als JSON speichern
summary_file = f"tutorial_summary_{timestamp}.json"
with open(summary_file, 'w', encoding='utf-8') as f:
    json.dump(summary, f, indent=2, ensure_ascii=False)

print(f"✅ Zusammenfassung erstellt: {summary_file}")
print(f"\n🎉 Alle Exports erfolgreich erstellt!")

# Zeige Zusammenfassung
print("\n📋 Tutorial-Zusammenfassung:")
print(f"  🤖 Agenten: {summary['agents_count']}")
print(f"  📝 Szenarien: {summary['scenarios_count']}")  
print(f"  ⚖️ Entscheidungen: {summary['total_decisions']}")
print(f"  📊 Validierungsscore: {summary['validation_score']:.1%}")
print(f"  📁 Erstellte Dateien: {len(summary['files_created'])}")
```
