## ğŸ¤– Schritt 1: Ethische Agenten erstellen

Agenten sind die Kernkomponenten unserer Simulation. Jeder Agent hat:
- **PersÃ¶nlichkeit**: Bestimmt Entscheidungstendenzen
- **Ethisches Framework**: Grundlegende ethische Prinzipien  
- **Lernmechanismen**: Wie der Agent aus Erfahrungen lernt
- **Soziale Eigenschaften**: Wie er mit anderen interagiert

### VerfÃ¼gbare PersÃ¶nlichkeitstypen:
- `utilitarian` - Maximiert Gesamtnutzen
- `deontological` - Folgt festen Regeln und Pflichten
- `virtue_ethics` - Strebt nach Tugenden und CharakterstÃ¤rke
- `balanced` - Ausgewogener Mix aller AnsÃ¤tze
- `pragmatic` - Praktische, situative Entscheidungen
- `idealistic` - Strebt nach hÃ¶chsten Idealen

```python
# ğŸ—ï¸ Erstelle eine diverse Gruppe von Agenten
print("ğŸ—ï¸ Erstelle ethische Agenten...")

# Verschiedene Agent-Konfigurationen
agent_configs = [
    {"id": "alice", "personality": "utilitarian", "framework": "utilitarian"},
    {"id": "bob", "personality": "deontological", "framework": "deontological"}, 
    {"id": "charlie", "personality": "virtue_ethics", "framework": "virtue_ethics"},
    {"id": "diana", "personality": "balanced", "framework": "balanced"},
    {"id": "eve", "personality": "pragmatic", "framework": "pragmatic"},
    {"id": "frank", "personality": "idealistic", "framework": "idealistic"}
]

# Agenten erstellen
agents = []
for config in agent_configs:
    agent = NeuralEthicalAgent(
        agent_id=config["id"],
        personality=config["personality"],
        ethical_framework=config["framework"]
    )
    agents.append(agent)
    print(f"âœ… {config['id']}: {config['personality']} Agent erstellt")

print(f"\nğŸ‰ {len(agents)} Agenten erfolgreich erstellt!")

# Zeige Agent-Details
for agent in agents:
    print(f"ğŸ¤– {agent.agent_id}: {agent.personality} ({agent.ethical_framework})")
    print(f"   Ãœberzeugungen: {len(agent.beliefs.beliefs)} Beliefs")
    print(f"   Entscheidungen: {len(agent.decision_history)}")
    print()
```

## ğŸ›ï¸ Schritt 2: Gesellschaft erstellen und Simulation durchfÃ¼hren

Die `NeuralEthicalSociety` koordiniert Interaktionen zwischen Agenten:
- **Soziale Netzwerke**: Dynamische Verbindungen zwischen Agenten
- **Gruppenbildung**: Automatische Community-Erkennung
- **Einflussverteilung**: Leadership und MeinungsfÃ¼hrerschaft
- **Lernprozesse**: Sozialer Einfluss und Wissensaustausch

```python
# ğŸ›ï¸ Gesellschaft erstellen
print("ğŸ›ï¸ Erstelle ethische Gesellschaft...")
society = NeuralEthicalSociety()

# Agenten zur Gesellschaft hinzufÃ¼gen  
for agent in agents:
    society.add_agent(agent)
    print(f"â• {agent.agent_id} zur Gesellschaft hinzugefÃ¼gt")

print(f"âœ… Gesellschaft mit {len(society.agents)} Agenten erstellt")

# ğŸ¬ Szenarien generieren und durchfÃ¼hren
print("\nğŸ¬ Generiere ethische Szenarien...")
scenario_gen = ScenarioGenerator()

# Mix aus vordefinierten und zufÃ¤lligen Szenarien
scenarios = [
    scenario_gen.get_trolley_problem(),
    scenario_gen.get_autonomous_vehicle_dilemma(), 
    scenario_gen.get_privacy_vs_security_dilemma()
]

# ZusÃ¤tzliche zufÃ¤llige Szenarien
for i in range(7):
    scenarios.append(scenario_gen.generate_random_scenario())

print(f"ğŸ“ {len(scenarios)} Szenarien generiert")

# Simulation durchfÃ¼hren
print("\nğŸš€ Starte Simulation...")
for i, scenario in enumerate(scenarios, 1):
    print(f"  ğŸ“ Scenario {i}/{len(scenarios)}: {scenario.title}")
    decisions = society.run_scenario(scenario)
    print(f"    âœ… {len(decisions)} Entscheidungen getroffen")

print("ğŸ‰ Simulation abgeschlossen!")
```

## ğŸ“Š Schritt 3: Umfassende Metriken sammeln

Das Metriken-System analysiert verschiedene Aspekte der Simulation:

### ğŸ›ï¸ Gesellschaftliche Metriken:
- **Polarisierung**: Wie stark gespalten sind die Meinungen?
- **Konsens**: Wie einig sind sich die Agenten? 
- **Netzwerk-KohÃ¤sion**: Wie zusammenhÃ¤ngend ist das soziale Netzwerk?
- **Einflussverteilung**: Wer hat wie viel Einfluss?

### ğŸ¤– Agenten-Metriken:
- **Interne Konsistenz**: Sind die Ãœberzeugungen widerspruchsfrei?
- **Zeitliche StabilitÃ¤t**: Wie stabil sind Ãœberzeugungen Ã¼ber Zeit?
- **EntscheidungsqualitÃ¤t**: Wie selbstsicher und durchdacht sind Entscheidungen?
- **Lernfortschritt**: Wie effektiv lernt der Agent?

```python
# ğŸ“Š Metriken sammeln
print("ğŸ“Š Sammle umfassende Metriken...")
metrics_collector = MetricsCollector()
all_metrics = metrics_collector.collect_all_metrics(society)

print("ğŸ“ˆ Gesellschaftliche Metriken:")
societal_metrics = all_metrics.get('societal_metrics', {})
for metric, value in societal_metrics.items():
    if isinstance(value, (int, float)):
        print(f"  ğŸ›ï¸ {metric}: {value:.3f}")
    else:
        print(f"  ğŸ›ï¸ {metric}: {value}")

print("\nğŸ‘¥ Agenten-Metriken (Durchschnitt):")
agent_metrics = all_metrics.get('agent_metrics', {})
if agent_metrics:
    # Berechne Durchschnittswerte
    avg_metrics = {}
    for agent_id, metrics in agent_metrics.items():
        for metric, value in metrics.items():
            if isinstance(value, (int, float)):
                if metric not in avg_metrics:
                    avg_metrics[metric] = []
                avg_metrics[metric].append(value)
    
    for metric, values in avg_metrics.items():
        avg_value = np.mean(values)
        print(f"  ğŸ¤– {metric}: {avg_value:.3f}")

print(f"\nâœ… {len(all_metrics)} Kategorien von Metriken gesammelt!")

# Zeige SchlÃ¼ssel-Insights
print("\nğŸ” SchlÃ¼ssel-Insights:")
polarization = societal_metrics.get('polarization', 0)
consensus = societal_metrics.get('consensus', 0) 
cohesion = societal_metrics.get('network_cohesion', 0)

print(f"ğŸ“ˆ Polarisierung: {polarization:.3f} ({'Niedrig' if polarization < 0.3 else 'Hoch' if polarization > 0.7 else 'Mittel'})")
print(f"ğŸ¤ Konsens: {consensus:.3f} ({'Niedrig' if consensus < 0.3 else 'Hoch' if consensus > 0.7 else 'Mittel'})")
print(f"ğŸ•¸ï¸ Netzwerk-KohÃ¤sion: {cohesion:.3f} ({'Schwach' if cohesion < 0.3 else 'Stark' if cohesion > 0.7 else 'Mittel'})")
```

## ğŸ” Schritt 4: Validierung und QualitÃ¤tssicherung

Das Validierungssystem Ã¼berprÃ¼ft:
- **Agent-Konsistenz**: Sind Ãœberzeugungen und Entscheidungen logisch konsistent?
- **Gesellschafts-StabilitÃ¤t**: Ist das soziale Netzwerk stabil und realistisch?
- **EntscheidungsplausibilitÃ¤t**: Sind die Entscheidungen nachvollziehbar?
- **LernvaliditÃ¤t**: Lernen die Agenten in sinnvoller Weise?
- **Anomalie-Erkennung**: Gibt es ungewÃ¶hnliche oder problematische Muster?

```python
# ğŸ” Validierung durchfÃ¼hren
print("ğŸ” FÃ¼hre umfassende Validierung durch...")
validator = ValidationSuite()
validation_results = validator.validate_society(society)

# Zeige Validierungsergebnisse  
print("ğŸ“‹ Validierungsergebnisse:")
passed = validation_results.get('passed_checks', 0)
total = validation_results.get('total_checks', 0) 
score = validation_results.get('overall_score', 0)

print(f"  âœ… Bestandene Tests: {passed}/{total}")
print(f"  ğŸ“Š Gesamtscore: {score:.1%}")

# Detaillierte Ergebnisse nach Kategorie
results_by_type = validation_results.get('results_by_type', {})
print("\nğŸ“‚ Ergebnisse nach Validierungstyp:")
for val_type, type_results in results_by_type.items():
    type_passed = type_results.get('passed', 0)
    type_total = type_results.get('total', 0)
    type_score = type_results.get('average_score', 0)
    print(f"  ğŸ”¸ {val_type}: {type_passed}/{type_total} ({type_score:.1%})")

# Zeige Warnungen und Probleme
all_results = validation_results.get('all_results', [])
warnings = [r for r in all_results if r.get('severity') == 'warning']
errors = [r for r in all_results if r.get('severity') == 'error']

if warnings:
    print(f"\nâš ï¸ Warnungen ({len(warnings)}):")
    for warning in warnings[:3]:  # Nur erste 3 zeigen
        print(f"  âš ï¸ {warning.get('check_name', 'Unknown')}: {warning.get('message', 'No message')}")
    if len(warnings) > 3:
        print(f"  ... und {len(warnings) - 3} weitere Warnungen")

if errors:
    print(f"\nâŒ Fehler ({len(errors)}):")
    for error in errors[:3]:  # Nur erste 3 zeigen
        print(f"  âŒ {error.get('check_name', 'Unknown')}: {error.get('message', 'No message')}")
    if len(errors) > 3:
        print(f"  ... und {len(errors) - 3} weitere Fehler")

print(f"\nâœ… Validierung abgeschlossen! Score: {score:.1%}")
```

## ğŸ’¾ Schritt 6: Datenexport und Reporting

Das Export-System unterstÃ¼tzt verschiedene Formate fÃ¼r weitere Analyse:
- **JSON**: Strukturierte Daten fÃ¼r Programmierung
- **CSV**: Tabellendaten fÃ¼r Excel/Pandas 
- **SQLite**: Relationale Datenbank fÃ¼r komplexe Abfragen
- **HTML**: Interaktive Berichte fÃ¼r PrÃ¤sentationen
- **Markdown**: Dokumentation und Reports

```python
# ğŸ’¾ Export und Reporting
print("ğŸ’¾ Erstelle Exports und Berichte...")
exporter = DataExporter(society)

# Timestamp fÃ¼r eindeutige Dateinamen
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

# 1. JSON Export (strukturierte Daten)
json_file = f"tutorial_export_{timestamp}.json"
exporter.export_json(json_file)
print(f"âœ… JSON Export erstellt: {json_file}")

# 2. CSV Export (Tabellendaten)
csv_file = f"tutorial_export_{timestamp}.csv" 
exporter.export_csv(csv_file)
print(f"âœ… CSV Export erstellt: {csv_file}")

# 3. HTML Bericht (interaktiv)
reporter = AutomatedReporter(society)
html_file = f"tutorial_report_{timestamp}.html"
reporter.generate_html_report(html_file)
print(f"âœ… HTML Bericht erstellt: {html_file}")

# 4. Zusammenfassung erstellen
summary = {
    'simulation_timestamp': timestamp,
    'agents_count': len(society.agents),
    'scenarios_count': len(scenarios),
    'total_decisions': sum(len(agent.decision_history) for agent in society.agents),
    'key_metrics': {
        'polarization': societal_metrics.get('polarization', 0),
        'consensus': societal_metrics.get('consensus', 0),
        'network_cohesion': societal_metrics.get('network_cohesion', 0)
    },
    'validation_score': validation_results.get('overall_score', 0),
    'files_created': [json_file, csv_file, html_file]
}

# Zusammenfassung als JSON speichern
summary_file = f"tutorial_summary_{timestamp}.json"
with open(summary_file, 'w', encoding='utf-8') as f:
    json.dump(summary, f, indent=2, ensure_ascii=False)

print(f"âœ… Zusammenfassung erstellt: {summary_file}")
print(f"\nğŸ‰ Alle Exports erfolgreich erstellt!")

# Zeige Zusammenfassung
print("\nğŸ“‹ Tutorial-Zusammenfassung:")
print(f"  ğŸ¤– Agenten: {summary['agents_count']}")
print(f"  ğŸ“ Szenarien: {summary['scenarios_count']}")  
print(f"  âš–ï¸ Entscheidungen: {summary['total_decisions']}")
print(f"  ğŸ“Š Validierungsscore: {summary['validation_score']:.1%}")
print(f"  ğŸ“ Erstellte Dateien: {len(summary['files_created'])}")
```
